```python
import numpy as np
import matplotlib.pyplot as plt
from math import e
```

```python
k = 1.380649e-23
h = 6.62607015e-34
c = 299792458
b = 2.9e-3
```

```python
def Planck(long, T):
  return (2 * h * c**2 / long**5) / (np.exp(h * c / (long * k * T)) - 1)
```

```python
x_1 = np.linspace(1e-6,300e-6,500)
x_2 = np.linspace(100e-9,3e-6,500)
x_3 = np.linspace(1e-9,500e-9,500)
T = np.array([100,5800,120000])
I_1 = Planck(x_1,T[0])
I_2 = Planck(x_2,T[1])
I_3 = Planck(x_3,T[2])
```

```python
#MÃ¡ximos
lamb_1 = (1e-9 * b / T[0]).item()
lamb_2 = (1e-9 * b / T[1]).item()
lamb_3 = (1e-9 * b / T[2]).item()
```

```python
fig, ax = plt.subplots(3,1, figsize = (7,8))
ax[0].plot(x_1*1e-9,I_1, label = 'T = 100 K', color = 'red')
ax[0].set_xlabel(r'Longitud de onda (nm)')
ax[0].set_ylabel(r'Intensidad ($\text{W}/\text{m}^3$)')
ax[0].axvline(x=lamb_1, label = fr'$\lambda = {lamb_1:.2e}$ nm')
ax[0].legend()

ax[1].plot(x_2*1e-9,I_2, label = 'T = 5800 K', color = 'blue')
ax[1].set_xlabel(r'Longitud de onda (nm)')
ax[1].set_ylabel(r'Intensidad ($\text{W}/\text{m}^3$)')
ax[1].axvline(x=lamb_2, label = fr'$\lambda_ = {lamb_2:.2e}$ nm')
ax[1].legend()

ax[2].plot(x_3*1e-9,I_3, label = 'T = 120000 K', color = 'green')
ax[2].set_xlabel(r'Longitud de onda (nm)')
ax[2].set_ylabel(r'Intensidad ($\text{W}/\text{m}^3$)')
ax[2].axvline(x=lamb_3, label = fr'$\lambda = {lamb_3:.2e}$ nm')
ax[2].legend()


plt.tight_layout()
plt.show()
```
